# Максимум и его распределение

Пусть $X_1, \ldots, X_n$ -- независимые (хотя бы попарно) и одинаково распределенные случайные величины. Вычислить функцию распределения
$$
\xi_1 = \max(X_1, \ldots, X_n), \quad \xi_2 = \min(X_1, \ldots, X_n).
$$

### Решение

По определению $F_{\xi}(x) = \mathsf{P}(\xi < x)$, тогда

$$
F_{\xi_1}(x) = \mathsf{P}(\max(X_1, \ldots, X_n) < x).
$$

Это выполняется в случае, когда все $X_i < x, i\in \{1,\dots, n\}$, откуда, с учетом независимости случайных величин и того, что они одинаково распределены, получаем

$$
F_{\xi_1}(x) = \mathsf{P}(X_1 < x, \dots, X_n < x) = F_{X_1}^n(x).
$$


Аналогично

$$
F_{\xi_2}(x) = \mathsf{P}(\min(X_1, \ldots, X_n) < x).
$$

Это условие выполняется, когда хотя бы одна случайная величина $X_i < x, i\in \{1,\dots, n\}:$

$$
F_{\xi_2}(x) = 1 - \mathsf{P}(X_1 \geq x, \dots, X_n \geq x).
$$

Используя независимость, и то, что все $X_i$ распределены одинаково, получаем

$$
\mathsf{P}(X_1 \geq x, \dots, X_n \geq x) =
$$

$$
=(1 - \mathsf{P}(X_1<x))\cdot \dots \cdot (1 - \mathsf{P}(X_n<x)) = (1 - F_{X_1}(x))^n.
$$

Откуда

$$
F_{\xi_2}(x) = 1 - (1 - F_{X_1}(x))^n
$$

# Свертки

## Напоминание

Пусть $\xi_1$ и $\xi_2$ — независимые случайные величины.

1. Если $\xi_1$ и $\xi_2$ имеют абсолютно непрерывное распределение, то

$$
f_{\xi_1 + \xi_2}(x) = \int\limits_{\mathbb{R}} f_{\xi_1}(t)f_{\xi_2}(x-t)dt.
$$

2. Если $\xi_1$ имеет дискретное распределение, а $\xi_2$ — абсолютно непрерывное распределение, то результатом суммы будет случайная величина, имеющая абсолютно непрерывное распределение и

$$
f_{\xi_1 + \xi_2}(x) = \sum\limits_{i=1}^{\infty} \mathsf{P}(\xi_1 = a_i)f_{\xi_2}(x-a_i).
$$

3. Если $\xi_1$ и $\xi_2$ имеют дискретное распределение, то

$$
 \mathsf{P}(\xi_1 + \xi_2 = b)= \sum\limits_{i=1}^{\infty} \mathsf{P}(\xi_1 = a_i)\mathsf{P}(\xi_2 = b-a_i).
$$

## Задача 1

Пусть случайные величины $\xi_1$ и $\xi_2$ независимы и их распределения задаются таблицами:

$$
\begin{array}{c|c|c}
\xi_1 & 1 & 2 \\
\hline
\mathsf P & \frac{1}{2} & \frac{1}{2}
\end{array}
$$


$$
\begin{array}{c|c|c|c}
\xi_2 & 0 & 1 & 2\\
\hline
\mathsf P & \frac{1}{3} & \frac{1}{3} & \frac{1}{3}
\end{array}
$$

Найти распределение случайной величины $\eta = \xi_1 + \xi_2$.

### Решение

Понятно, что $\eta$ будет принимать значения $\{1,2,3,4\}$.

1. $\mathsf{P}(\eta = 1)$:
$$
\mathsf{P}(\eta = 1) = \sum\limits_{i=1}^2\mathsf{P}(\xi_1 = a_i)\mathsf{P}(\xi_2 = 1 - a_i) =
$$

$$
= \mathsf{P}(\xi_1 = 1)\mathsf{P}(\xi_2 = 1 - 1) + \mathsf{P}(\xi_1 = 2)\mathsf{P}(\xi_2 = 1 - 2)
$$

$$
= \frac{1}{2}\cdot\frac{1}{3} + \frac{1}{2}\cdot0 = \frac{1}{6}.
$$

2. $\mathsf{P}(\eta = 2)$:
$$
\mathsf{P}(\eta = 2) = \mathsf{P}(\xi_1 = 1)\mathsf{P}(\xi_2 = 2 - 1) + \mathsf{P}(\xi_1 = 2)\mathsf{P}(\xi_2 = 2 - 2)
$$

$$
= \frac{1}{2}\cdot\frac{1}{3} + \frac{1}{2}\cdot\frac{1}{3} = \frac{1}{3}.
$$


3. $\mathsf{P}(\eta = 3)$:
$$
\mathsf{P}(\eta = 3) = \mathsf{P}(\xi_1 = 1)\mathsf{P}(\xi_2 = 3 - 1) + \mathsf{P}(\xi_1 = 2)\mathsf{P}(\xi_2 = 3 - 2)
$$

$$
= \frac{1}{2}\cdot\frac{1}{3} + \frac{1}{2}\cdot\frac{1}{3} = \frac{1}{3}.
$$

4. $\mathsf{P}(\eta = 4)$:
$$
\mathsf{P}(\eta = 4) = \mathsf{P}(\xi_1 = 1)\mathsf{P}(\xi_2 = 4 - 1) + \mathsf{P}(\xi_1 = 2)\mathsf{P}(\xi_2 = 4 - 2)
$$

$$
= \frac{1}{2}\cdot0 + \frac{1}{2}\cdot\frac{1}{3} = \frac{1}{6}.
$$

**Итого:**

$$
\begin{array}{c|c|c|c|c}
\eta & 1 & 2 & 3 & 4\\
\hline
\mathsf P & \frac{1}{6} & \frac{1}{3} & \frac{1}{3} & \frac{1}{6}
\end{array}
$$

## Задача 2

Пусть $\xi_1 \sim \mathsf{Exp}_{1}$, $\xi_2 \sim \mathsf{U}_{0,1}$, причем случайные величины независимы. Определить распределение случайной величины $\eta = \xi_1 + \xi_2$.

### Решение

Так как
$$
f_{\xi_1}(x) = \begin{cases} 0, & x < 0 \\ e^{-x}, & x\geq 0 \end{cases}.
$$
и
$$
f_{\xi_2}(x) = \begin{cases} 0, & x \notin [0,1] \\ 1, & x \in [0,1] \end{cases},
$$
то
$$
f_{\eta}(x) = \int\limits_{0}^\infty e^{-t}f_{\xi_2}(x-t)\,dt.
$$


#### Краткое решение

Перед нами полоса
$$
t \geq 0, \quad 0 \leq x - t \leq 1.
$$
В ней при $x \in [0, 1]$ выполняется $t \in [0, x]$ и
$$
f_\eta(x) = \int\limits_0^xe^{-t}\, dt = 1 - e^{-x}.
$$
Если $x > 1$, то $t \in [x-1, x]$ и
$$
f_\eta(x) = \int\limits_{x-1}^xe^{-t}\, dt = e^{1-x} - e^{-x}
$$

#### Подробное решение

Тогда для $\eta$

$$
f_{\eta}(x) = \int\limits_{\mathbb{R}} f_{\xi_1}(t)f_{\xi_2}(x-t)dt.
$$

Учитывая то, что $\xi_1$ имеет  показательное распределение, $f_{\xi_1}(t)$ будет отлично от $0$, только если

$$
t \geq 0.
$$

Аналогично, учитывая характер распределения $\xi_2$, получим, что $f_{\xi_2}(x-t) \neq 0$, только если

$$
0 \leq x-t \leq 1 \Longrightarrow x-1 \leq t \leq x.
$$

Тогда плотность случайной величины $\eta$ будет отлична от $0$, только если оба этих условия соблюдены.

Имеет смысл рассмотреть следующие варианты:

1. $x-1 \leq 0 \leq x \Longrightarrow x \in [0,1]$.

1-removebg-preview11.png

Тогда

$$
f_{\eta}(x) = \int\limits_{0}^{x} e^{-t}dt = e^{-t}|_{x}^{0} = 1-e^{-x}.
$$

2. $0 \leq x-1 \leq x \Longrightarrow x > 1$.

2-removebg-preview222.png

Тогда

$$
f_{\eta}(x) = \int\limits_{x-1}^{x} e^{-t}dt = e^{-t}|_{x}^{x-1} = e^{1-x}-e^{-x}.
$$

**Итого**

$$
f_{\eta}(x) = \begin{cases} 1-e^{-x}, & 0 \leq x \leq 1 \\ e^{1-x}-e^{-x}, & x > 1 \end{cases}.
$$

**Замечание**

Задачу также можно решить, положив

$$
f_{\eta}(x) = \int\limits_{\mathbb{R}} f_{\xi_2}(t)f_{\xi_1}(x-t)dt.
$$

Тогда отличные от $0$ значения плотности $\eta$ будут получены при

$$
t \in [0,1]
$$

и

$$
x-t \geq 0 \Longrightarrow t \leq x.
$$

## Задача 3

Пусть $\xi_1 \sim \mathsf{U}_{0,1}$, $\xi_2 \sim \mathsf{U}_{1,3}$. Определить распределение случайной величины $\eta = \xi_1 + \xi_2$.

### Решение

Запишем функции плотности для $\xi_1$ и $\xi_2$:

$$
f_{\xi_1}(x) = \begin{cases} 0, & x \notin [0,1] \\ 1, & x \in [0,1] \end{cases},
$$

$$
f_{\xi_2}(x) = \begin{cases} 0, & x \notin [1,3] \\ \frac{1}{2}, & x \in [1,3] \end{cases}.
$$

Для $\eta = \xi_1 + \xi_2$

$$
f_{\eta}(x) = \int\limits_{\mathbb{R}} f_{\xi_1}(t)f_{\xi_2}(x-t)dt.
$$

Плотность $\eta$ будет отлична от $0$ при условиях, что
$$
0 \leq t \leq 1
$$

и

$$
1\leq x-t \leq 3 ⟹ x-3 \leq t \leq x-1
$$

Учитывая, что длина отрезка $[x-3, x-1]$ равна $2$, имеет смысл рассмотреть 3 случая.

IMG_7125-removebg-preview.png

1. $x-3 < 0, \quad x-1 < 1, \quad x-1>0 \Longrightarrow x \in [1,2]$.

$$
f_{\eta}(x) = \int\limits_{0}^{x-1} \frac{1}{2}dt = \frac{x-1}{2}.
$$

2. $x-3 < 0, \quad x-1 > 1, \quad x-1>0 \Longrightarrow x \in [2,3]$.

$$
f_{\eta}(x) = \int\limits_{0}^{1} \frac{1}{2}dt = \frac{1}{2}.
$$

3. $x-3 > 0, \quad x-3 < 1, \quad x-1>1 \Longrightarrow x \in [3,4]$.

$$
f_{\eta}(x) = \int\limits_{x-3}^{1} \frac{1}{2}dt = \frac{4-x}{2}.
$$

**Итого**

Функция плотности для $\eta$ (собрать из кусочков, не забыть про 0).

# Ковариация

## Задача 1

Случайные величины $\xi$ и $\eta$ заданы таблицей совместного распределения:
$$
\begin{array}{  c | c | c | c  }
\eta \backslash \xi & 2 & 3 & 5  \\ \hline
-1 & 0.1 & 0.3 & 0.2 \\ \hline
1 & 0.1 & 0.05 & 0 \\ \hline
4 & 0 & 0.15 & 0.1
\end{array}
$$

Определить ковариацию $\mathsf{cov}(\xi, \eta)$.

Определить коэффициент корреляции $\rho(\xi, \eta)$.

### Решение

Ковариация двух случайных величин находится как:

$$
\mathsf{cov}(\xi, \eta) = \mathsf{E}(\xi\eta) - \mathsf{E}\xi\mathsf{E}\eta.
$$

Составим таблицу распределения для случайной величины $\xi\eta$. Для этого необходимо рассмотреть все возможные произведения значений случайных величин. Вероятность произведения для конкретного случая будет находиться на пересечении соответствующего столбца и строки. Если значения произведения получаются одинаковыми (в нашем случае такого нет), то необходимо сложить соответствующие вероятности. Значения, для которых вероятность равна $0$ в таблицу можно не заносить

$$
\begin{array}{  c | c | c | c | c | c | c | c }
\xi\eta & -5 & -3 & -2 & 2 & 3 &  12 & 20  \\ \hline
\mathsf{P} &0.2 & 0.3 & 0.1 & 0.1 & 0.05 & 0.15 & 0.1
\end{array}
$$
    
Составим таблицы маргинальных распределений $\xi$ и $\eta$:

$$
\begin{array}{  c | c | c | c  }
\xi & 2 & 3 & 5   \\ \hline
\mathsf{P} &0.2 & 0.5 & 0.3
\end{array}
$$

$$
\begin{array}{  c | c | c | c  }
\eta & -1 & 1 & 4   \\ \hline
\mathsf{P} &0.6 & 0.15 & 0.25
\end{array}
$$

Найдем соответствующие математические ожидания:

$$
\mathsf{E}(\xi\eta) = 2.05,
$$

$$
\mathsf{E}\xi = 3.4,
$$

$$
\mathsf{E}\eta = 0.55.
$$

Тогда

$$
\mathsf{cov}(\xi, \eta) = 2.05 - 3.4\cdot0.55 = 0.18.
$$

Найдем коэффициент корреляции $\rho$ для наших случайных величин. Напомним, что

$$
\rho(\xi,\eta) = \frac{\mathsf{cov}(\xi, \eta)}{\sqrt{\mathsf{D}\xi}\cdot\sqrt{\mathsf{D}\eta}}
$$

Найдем соответствующие дисперсии $\xi$ и $\eta$ и подставим в выражение для $\rho$ :

$$
\rho(\xi,\eta) = \frac{0.18}{\sqrt{1.24}\cdot\sqrt{4.4475}} \approx 0.077
$$

# Многомерные распределения

## Задача 1

Двумерная непрерывная случайная величина распределена равномерно в области $K: x\in[1,3], y \in [2,4]$

Составить функцию плотности $f_{\vec{\xi}}(x,y)$:

*   $(x, y) \notin K $
*   $(x, y) \in K $

Составить одномерную функцию плотности $f_{\xi_{1}}(x)$:

*   $x \notin [1,3] $
*   $x \in [1,3]$

Составить одномерную функцию плотности $f_{\xi_{2}}(y)$
*   $y \notin [2,4] $
*   $y \in [2,4]$


Найти $\mathsf{P}(2< x< 3, 1 \leq y \leq 3)$

Сделать вывод о зависимости величин $\xi_1$ и $\xi_2$

### Решение

Так как область $K$ представляет собой квадрат со стороной $2$, то функция плотности будет иметь вид:

$$
 f_{\vec{\xi}}(x,y) = \begin{cases}
  0, & (x, y) \notin K
   \\
   \frac{1}{4},  & (x, y)\in K.
\end{cases}
$$

Первая компонента плотности:

$$
f_{\xi_{1}}(x) = \int\limits_{-\infty}^{+\infty}f(x,y)dy = \frac{1}{4}\int\limits^4_2 dy = \frac{1}{2},
$$

если $x\in [1,3]$, иначе $0$.<br/>

Вторая компонента плотности:

$$
f_{\xi_{2}}(y) = \int\limits_{-\infty}^{+\infty}f(x,y)dx = \frac{1}{4}\int\limits^3_1 dx = \frac{1}{2},
$$

  если $y\in [2,4]$, иначе $0$. <br/>

Найдем $\mathsf{P}(2< x < 3, 1 \leq y \leq 3)$:

$$
\mathsf{P}(2< x< 3, 1 \leq y \leq 3) = \int\limits^3_2 \int\limits^3_1 f_{\vec{\xi}}(x,y) dxdy  =  
$$

$$
=\int\limits^3_2 \int\limits^2_1 0 dxdy + \int\limits^3_2 \int\limits^3_2 \frac{1}{4} dxdy = \frac{1}{4}.
$$

Вероятность можно вычислить и из геометрических соображений: площадь пересечения прямоугольника, ограниченного $2 < x < 3, 1 \leq y \leq 3$ и области $K$ разделить на площадь области $K$.

4_rectangle.png

В случае, если $x \in [1,3], y \in [2,4]$:

$$
f_{\vec{\xi}}(x,y) = f_{\xi_{1}}(x)\cdot f_{\xi_{2}}(y) = \frac{1}{4}.
$$

Во всех остальных случаях:

$$
f_{\vec{\xi}}(x,y) = f_{\xi_{1}}(x)\cdot f_{\xi_{2}}(y) = 0.
$$

Таком образом можно сделать вывод, что случайные величины $\xi_1$ и $\xi_2$ независимы.
